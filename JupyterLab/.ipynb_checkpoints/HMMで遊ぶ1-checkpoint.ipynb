```python
import pandas as pd
import numpy as np
from hmmlearn import hmm
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
%matplotlib inline
```

```python
def random_transition(n=100, w_0=np.array([1.0, 0.0, 0.0]) ) :
    # return [sunny, rainy, cloudy]
    X = np.array(
        [[0.4, 0.1, 0.3],
         [0.1, 0.5, 0.3],
         [0.5, 0.4, 0.4]]
    )
    w = w_0[:]
    for i in range(n):
        w = X @ w
        w = np.eye(3)[np.random.choice(a=[0,1,2], p=w)]
        yield w
def random_action(w) :
    # return [walk, shop, cleen]
    if w.argmax() == 0 :
        return np.eye(3)[np.random.choice(a=[0,1,2], p=[.5, .1, .4])]
    if w.argmax() == 1 :
        return np.eye(3)[np.random.choice(a=[0,1,2], p=[.0, .7, .3])]
    if w.argmax() == 2 :
        return np.eye(3)[np.random.choice(a=[0,1,2], p=[.3, .4, .3])]
```

```python
w = np.array([w for w in random_transition(n=10000)])
y = np.array([*map(random_action, w)])
w_index = np.array([*map(np.argmax, w)])
y_index = np.array([*map(np.argmax, y)])
```

```python
plt.hist(w_index)
ax = plt.gca()
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(['Sunny','Rainy', 'Cloudy'])
```

```python
def only_one_state_y(s, w, y):
    index = [*map(lambda b:b[0],[*filter(lambda a: a[1]==s, [*enumerate([*map(np.argmax, w)])])])]
    return y[index]
w0_y_index = np.array([*map(np.argmax, only_one_state_y(0, w, y))])
w1_y_index = np.array([*map(np.argmax, only_one_state_y(1, w, y))])
w2_y_index = np.array([*map(np.argmax, only_one_state_y(2, w, y))])
```

```python
plt.hist(w1_y_index)
ax = plt.gca()
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(['Wolk','Shop', 'Clean'])
```

```python
X = np.array([*map(np.argmax, y)])
X = X.reshape(-1, 1)
print(X.shape)
# 行動から天気を三つに分類する
model = hmm.MultinomialHMM(n_components=3)
# startprob は不明
model.transmat = np.array(
                    [[0.4, 0.1, 0.3],
                     [0.1, 0.5, 0.3],
                     [0.5, 0.4, 0.4]]
                 )
model.emissionprob = np.array(
                        [[0.5, 0.1, 0.4],
                         [0.0, 0.7, 0.3],
                         [0.3, 0.4, 0.3]])
model.fit(X)
L,Z = model.decode(X)
```

```python
plt.plot(Z[:90])
plt.plot(w_index[:90])
```
