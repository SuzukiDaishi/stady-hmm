```python
import numpy as np
import matplotlib.pyplot as plt
from hmmlearn import hmm
%matplotlib inline
plt.style.use('ggplot')
```

```python
X = np.array([[np.random.binomial(1, p=0.5)] for i in range(100)])
plt.plot(X)
plt.yticks((0,1))
plt.xlabel('Trial')
ax = plt.gca()
ax.set_yticklabels(['Tail','Head'])
```

```python
X = np.array([[np.random.binomial(1, p=0.7)] for i in range(50)])
X = np.append(X, np.array([[np.random.binomial(1, p=0.0)] for i in range(50)]), axis=0)
plt.plot(X)
plt.xlabel('Trial')
plt.yticks((0,1))
ax = plt.gca()
ax.set_yticklabels(['Tail','Head'])
```

```python
model = hmm.MultinomialHMM(n_components=2)
model.fit(X)
L,Z = model.decode(X)
```

```python
plt.plot(Z)
plt.xlabel('Trial')
plt.yticks((0,1))
ax = plt.gca()
ax.set_yticklabels(['Fair','Cheating'])
```

```python
print('[[モデル1裏, モデル1表],\n [モデル1裏, モデル1表]]\n')
print(model.emissionprob_, '\n')
print('モデル1で表の確率: {:f},\nモデル２で表の確率: {:f}'.format(model.emissionprob_[0, 1], model.emissionprob_[1, 1]))
```
